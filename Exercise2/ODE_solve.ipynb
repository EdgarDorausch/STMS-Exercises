```python
from typing import *
import numpy as np
import matplotlib.pyplot as plt
import math

%matplotlib inline
```

```python
c1 = c2 = 1.0
xT = 2e5
x0 = 0.8e5

lim = c2*xT/(c1+c2)

def euler_solution(dt: float, T: float) -> Tuple[np.ndarray, np.ndarray]:
    t = np.arange(start=0.0, stop=T, step=dt)
    N = t.shape[0] # number of steps

    x = np.empty([N])
    x[0] = x0

    for i in range(N-1):
        xi = x[i]

        dx_dt = c2*xT - (c1+c2)*xi
        x[i+1] = xi + dx_dt*dt

    return x,t
```

```python
def plot_euler_solution(dt: float, T: float):
    x,t = euler_solution(dt, T)

    plt.hlines(lim, 0.0, T)
    plt.plot(t, x)
```

```python
T = 4.0

plot_euler_solution(0.1, T)
plot_euler_solution(0.2, T)
plot_euler_solution(0.3, T)
```

```python
plot_euler_solution(1.5, T)
plot_euler_solution(1.0, T)
plot_euler_solution(0.5, T)
```
